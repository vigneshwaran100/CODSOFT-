```python
import pandas as pd

df=pd.read_csv('/content/csv file/AB_NYC_2019.csv')
df
```

```python
df.describe()
```

```python
df_outlier=df.price.quantile(0.99)
df_new_data=df[df.price<df_outlier]
df_new_data
```

```python
df_new_data.describe()
```

```python
df_new_data.describe()
```

```python
df_min_outlier=df_new_data['price'].quantile(0.0)
df_data=df_new_data[df_new_data.price>df_min_outlier]
df_data
```
